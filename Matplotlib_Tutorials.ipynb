# Import Numpy and Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Generate Random Dataset

```python
X_data = np.random.random(50)*100
Y_data = np.random.random(50)*100
```

# Scatter Plots

```python
plt.scatter(X_data,Y_data, c='black',s=50,marker='*')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.show()
```

# Line Plots

```python
years = [2018 + x for x in range(9)]
weight = [58,60,70,72,69,75,72,74,66]
plt.plot(years,weight,c='black',linewidth =3)
plt.xlabel('Years')
plt.ylabel('Weight')
plt.title('Line Plot')
plt.show()
```

# Bar Plot


```python
Fruits = ['Mango','Guava','Banana','Apple','Grapes']
Votes = [100,30,45,77,32]
plt.bar(Fruits,Votes,color ='red',width=0.5)
plt.xlabel('Fruits')
plt.ylabel('Votes')
plt.title('Bar Plot')
plt.show()
```

# Histogram

```python
marks = np.random.normal(20,1.5,100)
plt.hist(marks,color='red')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# Pie Chart

```python
Fruits = ['Mango','Guava','Banana','Apple','Grapes']
Votes = [100,30,45,77,32]
explodes = (.01,.01,.01,.01,.01)
plt.pie(Votes,labels=Fruits,autopct='%1.1f%%',explode=explodes)
plt.title('Pie Chart')
plt.show()
```

# Box Plot

```python
heights = np.random.normal(180,5,100)
plt.boxplot(heights)
plt.title('Box Plot')
plt.show()
```

# Plot Customization

```python
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018]
income = [10,15,19,20,26,29,31,38,36]
income_tick = list(range(5,40,5))

plt.plot(years,income)
plt.title('Income Graph',fontsize = 20)
plt.xlabel('Years',fontsize = 10)
plt.ylabel('Income',fontsize = 10)
plt.yticks(income_tick,[f"{x} k USD" for x in income_tick])
plt.show()
```

# Legends and Multiple Plots

```python
tempday1 = [10,12,15,11,25]
tempday2 = [11,13,8,19,15]
tempday3 = [7,9,11,14,8]

plt.plot(tempday1,label = 'Day 1')
plt.plot(tempday2,label = 'Day 2')
plt.plot(tempday3,label = 'Day 3')
plt.title('Temperature Graph')
plt.legend()
plt.show
```

# Multiple Figures

```python
x1,y1 = np.random.random(50),np.random.random(50)
x2,y2 = np.arange(50),np.random.random(50)

plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)

plt.show
```
